```python
import sqlite3
```

```python
conn = sqlite3.connect('HW3_AFC_North_APPG.db')
```

```python
conn.execute('''
CREATE TABLE HW3_AFC_North_APPG(
  team_name TEXT PRIMARY KEY NOT NULL,
  coach TEXT,
  wins INTEGER,
  losses INTEGER,
  ties INTEGER)
  STRICT;
''')
```

```python
conn.execute('''
INSERT INTO HW3_AFC_North_APPG (team_name, coach, wins, losses, ties)
VALUES ('Steelers', 'Mike Tomlin', 10, 7, 0),
('Ravens', 'John Harbaugh', 8, 9, 0),
('Bengals', 'Zach Taylor', 6, 11, 0),
('Browns', 'Kevin Stefanski', 5, 12, 0);
''')
```

```python
query = "SELECT * FROM HW3_AFC_North_APPG WHERE wins > 7"
results = conn.execute(query)
for row in results:
  print(row)
```

```python
query = "SELECT team_name, wins FROM HW3_AFC_North_APPG ORDER BY wins DESC"
for row in conn.execute(query):
  print(row)
```
